# read_csv
- *read_csv():* para ler arquivos CSV e outros formatos de dados estruturados
- *sep:* separador de campo usado no arquivo CSV (,; |)
- *decimal:* caractere de separação de casa decimal (,.)


```python
# exemplo
import pandas as pd
df = pd.read_csv("Cancer_Data.csv")
df.head() # mostra as 5 primeiras linhas da planilha
```

```python
# exemplo usando o sep e o decimal 
import pandas as pd
df = pd.read_csv("Cancer_Data.csv",
                 sep = ',',
                 decimal = '.')
df.head()
```

# Index
É uma sequência de rótulos que *identifica exclusivamente cada linha do DataFrame*. O índice pode ser *numérico*, como uma sequência de arrays inteiros, ou pode ser uma *sequência de rótulos*, como uma lista de strings

```python
print(df.index) 
print(list(df.index))
```

O método *set_index()* define uma ou mais colunas existentes como o novo índice DataFrame
O método *reset_index()* remove o índice atual e cria um novo índice numérico começando do zero

```python
nomes_df = pd.DataFrame({
    'nome': ['Alice', 'Bob', 'Larissa'],
    'idade': [15, 30, 26]
})

nomes_df

#testando o set_index():
nomes_df_new = nomes_df.set_index("nome")
nomes_df_new

# resetando o index para o anterior com o reset_index():
nomes_df_back = nomes_df_new.reset_index()
nomes_df_back
```

# Columns
- O atributo colums retorna o *nome E o tipo das colunas* do dataframe
- Utilize *colchetes* para criar novas colunas

# Filtros
Filtrar dados com base em uma condição *booleana*. A condição é uma série booleana que tem o mesmo comprimento que o dataframe e é usada para selecionar linhas que satisfazem a mesma *condição*


```python
#exemplo
filtro = nomes_df['idade'] > 20 # o filtro será na coluna idade cujos valores sejam maior que 20
nomes_filtrado = nomes_df[filtro]
nomes_filtrado

# também é possível escrever dessa forma
nomes_df[nomes_df['idade'] > 26]
```

# loc
É usado para *acessar linhas e colunas* em um dataframe usando *rótulos*. O método aceita dois parâmetros: 
1) Especifica as linhas que devem ser acessadas
2) Especifica as colunas que devem ser acessadas

```python
# exemplo loc
nomes_df.loc[1, 'nome'] # retorna o nome na linha de índice 1

# exemplo loc com filtro
nomes_df.loc[nomes_df['idade'] > 20, 'nome'] # retorna APENAS os nomes das pessoas que possuem idade maior que 20
```

# iloc
É usado para acessar linhas e colunas ocultas em um dataframe usando índices inteiros. O método aceita dois parâmetros:
1) Especifica as linhas acessadas
2) Colunas que devem ser acessadas

Tanto o *loc* quanto o *iloc* podem ser usados para fazer seleções de linhas e colunas em um dataframe, mas a diferença é que o primeiro usa *rótulos* e o segundo usa *índices inteiros*

```python
#exemplo iloc
nomes_df.iloc[1, 0] #linha 1 coluna 0
```

```python
# exemplo iloc
nomes_df.iloc[[0, 1], 0] # duas linhas na coluna 0
#  está especificando que queremos acessar as linhas de índice 0 e 1, e o número 0 indica que estamos interessados na primeira coluna ('nome') do DataFrame.
```

- média

# missing values (valores ausentes)
Referenm-se a valores que não estão presentes em um conjunto de dados, geralmente indicados por um valor especial como *NaN* (not a number) ou *None*.
Os valores ausentes podem ocorrer por vários motivos, como erros de coleta de dados, problemas técnicos ou simplesmente porque o valor é desconhecido.

## isnull()
É usada para identificar valores ausentes em um DataFrame e retorna um DF booleano do mesmo tamanho do DataFrame original, onde cada elemento é "True" se o valor for ausente e "False" caso o contrário.

```python
# exemplo
fastfood[fastfood['protein'].isnull()] #filtrando itens que possuem valores null no campo protein

# exemplo 2
fastfood.isnull() #verifica se existem campos null em cada item
```

# dropna()
Elimina linhas ou colunas inteiras que possuam valores ausentes.

```python
# removendo linhas com valores ausentes no campo protein
print(f"antes: {fastfood.shape}")
fastfood = fastfood.dropna(subset= 'protein')
print(f"depois: {fastfood.shape}")
```

```python
# removendo todas as linhas que contém valores vazios
fastfood = pd.read_excel("fastfood.xlsx")
print(f"antes: {fastfood.shape}")

fastfood = fastfood.dropna()
print(f"depois: {fastfood.shape}")
```

# Operações com string
Essas funções são chamadas de métodos de string e podem ser usadas para realizar operações de manipulação de strings em uma ou mais colunas do DataFrame.

## str.len()
Retorna o comprimento de cada elemento da coluna de string.

## str.lower() e str.upper()
Converte todas as letras em minúsculas ou maiúsculas, respectivamente.


```python
# exemplo str len
fastfood['item'].str.len() # tamanho das palavras de cada elemento da coluna item

#lower
fastfood['item'].str.lower() # transforma todas as palavras da coluna em minusculas

#upper
fastfood['item'].str.upper() # transforma todas as palavras da coluna em maiusculas
```

## str.replace()
Substitui um valor por outro em uma coluna de strings

## str.contains()
Verifica se uma string específica está presente em uma coluna de string e retorna uma série booleana

## str.split()
Divide uma string em substrings com base em um separador e retorna uma série de listas

```python
# verificando quais elementos da coluna item contém a palavra burger
fastfood['item'].str.contains("Burger")
```

```python
# substituindo a palavra burger para xburger na coluna item
fastfood['item'].str.replace("Burger", "XBurger")
```

```python
# dividindo strings na coluna item que tiverem espaço entre as palavras
fastfood['item'].str.split(" ")
```
