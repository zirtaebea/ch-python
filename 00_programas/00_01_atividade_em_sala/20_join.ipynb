```python
# 1) Aplique as quatro formas de join estudadas nos seguintes dataframes

# importando bibliotecas 
import pandas as pd

# criando dataframe
df1 = pd.DataFrame({'A': [4, 3, 2, 1], 'B': ['a', 'b', 'c', 'b']})
df2 = pd.DataFrame({'A': [2, 3, 5], 'C': [10, 20, 30]})
```

```python
# fazendo joins
# inner join
df1.merge(df2, on='A', how='inner')
```

```python
# fazendo joins
# left join
df1.merge(df2, on='A', how='left')
```

```python
# fazendo joins
# right join
df1.merge(df2, on='A', how='right')
```

```python
# fazendo joins
# outer join
df1.merge(df2, on='A', how='outer')
```

```python
# 2) Sobre o resultado do outer join anterior, ordene a saída pela coluna A (ajustar índices). 

# outer join
df3 = df1.merge(df2, on='A', how='outer')
df3 = df3.sort_values(['A'])
df3.reset_index(drop=True) # para resetar o index
```

```python
# 3) Concatenar df1 e df2 em formato de linhas e depois colunas (ajustar índices).

# por linhas
df4 = pd.concat([df1, df2], ignore_index=True)
df4
```

```python
# por colunas
df5 = pd.concat([df1, df2], axis=1)
df5
```

```python
# calcule a soma da coluna A por agrupamento da coluna B no df1
df1.groupby('B').sum().reset_index()
```
