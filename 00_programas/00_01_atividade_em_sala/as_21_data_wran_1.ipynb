Utilizando a base "realtor-data.csv" (link kaggle) gerar uma amostra com os seguintes requisitos: 
* amostra aleatória
* com 1.000 registros, sendo:
* - 250 do estado de Massachusetts
* - 250 do estado de Puerto Rico
* - 250 do estado de Connecticut       
* - 250 dos estados Rhode Island e New York
* somente casas for_sale
* sem duplicidades de status, city, state e price
* na remoção de duplicados priorizar as casas maiores. 
Valide a amostra criada. 

```python
import pandas as pd
realtor = pd.read_csv('realtor-data.csv')
realtor
```

```python
# analisando dados
# planilha geral
realtor.info()
```

```python
#verificando quantidades
# de casas for sale
realtor.groupby('status')[['price']].count()
```

```python
# de estados
realtor.groupby('state')[['price']].count()
```

```python
# apenas casas for sale
realtor = realtor[realtor['status'] == 'for_sale']
realtor
```

```python
# filtrando os estados
# Massachusetts	
massachu = realtor[realtor['state'] == 'Massachusetts']

# Puerto Rico
puerto = realtor[realtor['state'] == 'Puerto Rico']

# Connecticut
connectcut = realtor[realtor['state'] == 'Connecticut']

# Rhode Island e New York
rhode_ny = realtor[(realtor['state'] == 'Rhode Island') | (realtor['state'] == 'New York')]
```

```python
# ordenar do maior para o menor para que na remoção de duplicados sejam priorizadas as casas maiores.
massachu = massachu.sort_values('house_size', ascending=True)
puerto = puerto.sort_values('house_size', ascending=True)
connectcut = connectcut.sort_values('house_size', ascending=True)
rhode_ny = rhode_ny.sort_values('house_size', ascending=True)
```

```python
# sem duplicidades de status, city, state e price
massachu = massachu.drop_duplicates(subset=['status', 'city', 'state', 'price'], keep='last')
puerto = puerto.drop_duplicates(subset=['status', 'city', 'state', 'price'], keep='last')
connectcut = connectcut.drop_duplicates(subset=['status', 'city', 'state', 'price'], keep='last')
rhode_ny = rhode_ny.drop_duplicates(subset=['status', 'city', 'state', 'price'], keep='last')
rhode_ny
```

```python
# criando amostras
amostra_massachu = massachu.sample(n=250, random_state=100)
amostra_puerto = puerto.sample(n=250, random_state=100)
amostra_connecticut = connectcut.sample(n=250, random_state=100)
amostra_rhodeny = rhode_ny.sample(n=250, random_state=100)

amostra_massachu
```

```python
# unindo amostras
amostra_geral = pd.concat([amostra_massachu, amostra_puerto, amostra_connecticut, amostra_rhodeny], ignore_index=True)
amostra_geral
```

```python
# validando
validacao = amostra_geral.groupby(['status', 'state'])[['price']].count()
validacao
```
