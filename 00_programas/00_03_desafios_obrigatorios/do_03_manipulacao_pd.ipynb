```python
# importando biblioteca
import pandas as pd

consoles = pd.read_excel('consoles.xlsx')

# verificando df
consoles.head(10)

```

```python
# 1) fazendo contagem de quantos consoles foram lançados por cada companhia
contagem_console = consoles.groupby('Company')[['Console Name']].count()
contagem_console
```

```python
# 2) Ordene o resultado anterior pela quantidade de consoles por companhia 
# de forma decrescente. 

contagem_console = contagem_console.sort_values('Console Name', ascending=False)
contagem_console
```

```python

# 3) Sobre o dataframe anterior, adicione a coluna "Market Value" 
# presente na tabela do arquivo "company.txt".

# importando dataframe com as informações da coluna nova
mkt_value = pd.read_csv('company.txt')
mkt_value

# realizando o merge por companhia com 
# o INNER JOIN pois quero manter as o count do group by por console name

console_mkt = contagem_console.merge(mkt_value, on='Company', how='inner')
console_mkt
```

```python
# 4) Qual o valor médio de Market Value por console?
# somando todos os consoles
console_soma = console_mkt['Console Name'].sum()

# somando todos os mkt values
mkt_soma = console_mkt['Market Value'].sum()

# resultado
vm_console = mkt_soma/console_soma

# imprimindo 
print(f"O valor médio do Market Value por console é de: {vm_console.round(2)}")
```
