```python
# importando bibliotecas
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# importando excel 
fastfood = pd.read_excel('fastfood.xlsx')
# verificando importação do df
fastfood.head(10)
```

```python

# GRAFICO DE BARRAS
# dataframe de itens por restaurante
rest_item = fastfood.groupby('restaurant')[['restaurant']].count()

# renomeando a coluna do count para item
rest_item.columns = ['item']

# resetando os índices do df para correlacionar os dados das duas colunas corretamente
restaurantes = rest_item.reset_index()

# verificando 
restaurantes

# formatando o gráfico
# tamanho
plt.figure(figsize=(10, 10))
# gráfico
sns.barplot(x= 'restaurant', y='item', data=restaurantes)
# título
plt.title('Contagem de itens do cardápio por restaurante')
plt.xlabel('Restaurantes')
plt.ylabel('Itens do cardápio')
plt.show()
```

```python
# HISTOGRAMA E DISTRIBUIÇÃO
# pratos com hamburguer por restaurante
burger = fastfood.loc[fastfood['item'].str.contains("Burger") == True, 'restaurant' ]
burger = burger.reset_index(drop=True)

# formatando
plt.figure(figsize=(6,10))

# histograma
sns.displot(burger, kde=True)

# titulo do gráfico
plt.title('Número de pratos com hamburguer por restaurante')

# renomeando eixos
plt.xlabel('Restaurantes')
plt.ylabel('Pratos com hamburguer')

# mostrando resultados
plt.show()
```

```python
# GRAFICO DE BARRAS
# media de calorias por restaurante
media_cal = fastfood.groupby('restaurant')[['calories']].mean(numeric_only=True).round(1).reset_index()
media_cal

# gráfico de barras
plt.figure(figsize=(10, 10))
plt.bar(media_cal['restaurant'], media_cal['calories'], color='pink')

# editando rotulos
plt.xlabel('Restaurantes')
plt.ylabel('Média de calorias')
plt.title('Média de calorias por restaurante')

# mostrando resultados
plt.show()
```

```python
# dispersão 
# gorduras todais por calorias 
sns.scatterplot(x='calories', y='total_fat', data=fastfood)
plt.show()
```
