# Matplotlib
Pode ser usada tanto em ambientes interativos como em scripts Python, e é altamente integrável com outras bibliotecas de análise de dados, como NumPy e Pandas. É possível personalizar cada aspecto do gráfico, desde os rótulos dos eixos até a cor e o tamanho dos pontos.

Ela oferece uma ampla variedade de gráficos e gráficos, incluindo:
- gráficos de linha, 
- gráficos de dispersão, 
- gráficos de barras, 
- gráficos de área,
- histogramas


````python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
````

## Tipos de gráficos 
- **linha:** plt.plot(x, y)
  mostra a tendência ao longo do tempo ou do eixo x.

- **barras:** plt.bar(x, y)
  é utilizado para comparar dados em diferentes categorias.

- **pizza:** plt.pie(quantidade, labels=labels)
  mostra a distribuição percentual de dados em uma única categoria.

- **dispersão:** plt.scatter(x, y)
  mostra a relação entre duas variáveis numéricas.

- **histograma:** plt.hist(dados)
  mostra a distribuição de frequência de dados em um intervalo. (https://www.significados.com.br/histograma/#:~:text=Definir%20quantas%20classes%20(barras)%20ser%C3%A3o,e%20menor%20valor%20de%20intervalos).

- **área:**: plt.fill_between(x, y)
  mostra a tendência e a magnitude da mudança em dados ao longo do tempo.


````python
#barras
x = ['A', 'B', 'C', 'D', 'E']
y = [12, 15, 4, 7, 35]
plt.bar(x, y)
#mostrando resultados
plt.show()
````

````python
#pizza
labels = ['Maçã', 'Pera', 'Uva', 'Laranja']
quantidade = [36, 18, 18, 28]
plt.pie(quantidade, labels=labels)
# mostrando resultados
plt.show()
````

````python
#dispersão
x = [2, 4, 6, 8, 10]
y = [20, 18, 16, 14, 12] 
plt.scatter(x, y)
# mostrando resultados
plt.show()
````

````python
#histograma
dados = [1, 3, 3, 4, 4, 4, 5, 5, 6, 7, 8, 9]
plt.hist(dados, bins=4)
# mostrando resultados
plt.show()
````

````python
# area
x = [2, 4, 6, 8, 10]
y = [20, 18, 16, 14, 12] 
plt.fill_between(x, y)
# mostrando resultados
plt.show()
````

# Configurações
## Legendas
- **Título do gráfico:** definir o título do gráfico usando a função ``title()``
- **Etiquetas dos eixos:** definir as etiquetas dos eixos usando as funções ``xlabel()`` e ``ylabel()``
- **Limites dos eixos:** definir os limites dos eixos usando as funções ``xlim()`` e ``ylim()``
- **Escala dos eixos:** definir a escala dos eixos usando as funções ``xscale()`` e ``yscale()``
- **Legenda:** adicionar uma legenda aos seus gráficos usando a função ``legend()``
- **Cores:** definir a cor das linhas, barras ou pontos usando o argumento ``color`` ou ``c``
- **Estilo de linha:** você pode definir o estilo da linha usando o argumento ``linestyle`` ou ``ls``
- **Tamanho do marcador:** definir o tamanho dos pontos usando o argumento ``markersize`` ou ``ms``
- **Tipo de marcador:** definir o tipo de marcador usando o argumento ``marker``
- **Transparência:** definir a transparência das linhas, barras ou pontos usando o argumento ``alpha``
- **Espessura da linha:** definir a espessura da linha usando o argumento ``linewidth`` ou ``lw``
- **Grade:** adicionar ou remover uma grade no gráfico usando a função ``grid()``


````python
# importando biblioteca
import matplotlib.pyplot as plt

#dados
x = [2, 4, 6, 8, 10]
y = [10, 8, 6, 4, 2] 

#configurações do gráfico
plt.plot(x, y,
         color='red',
         ls = '--',
         lw = 3,
         marker='o',
         ms = 8,
         markerfacecolor='white'
         )
plt.title('Gráfico de teste')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.xlim(0, 12)
plt.ylim(0, 12)
plt.grid(True)

#mostrando resultados
plt.show()
````

# Subplots
O método ``subplot()`` do Matplotlib é usado para criar múltiplos gráficos em uma figura. Com o subplot(), você pode especificar o número de linhas e colunas de gráficos e a posição de cada gráfico na figura.


````python
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# 3 linhas e uma coluna, com o tamanho 8x10px
fig, axs = plt.subplots(3, 1, figsize=(8, 10))
axs[0].plot(x, y1) # criando primeiro gráfico
axs[0].set_title('Seno') # setando nome do gráfico
axs[1].plot(x, y2)
axs[1].set_title('Cosseno')
axs[2].plot(x, y3)
axs[2].set_title('Tangente')

# mostrando resultados
plt.show()
````

# Seaborn
O Seaborn é uma biblioteca de visualização de dados baseada no Matplotlib. 
Enquanto o Matplotlib é uma biblioteca mais básica e flexível, o Seaborn oferece uma interface mais simplificada e especializada para a criação de gráficos estatísticos atraentes e informativos.

## Gráficos de distribuição
### **Visualização de dados:** 
Podemos usar o ```displot()``` para a visualização de valores de uma variável

````python
import seaborn as sns
x = np.random.normal(size=100)

# plotar histograma e distribuição
sns.displot(x, kde=True)
plt.show()
````

### **Visualização de relações entre variáveis:**
Podemos usar o ``scatterplot()`` para visualizar a relação entre duas variáveis.


````python
# importando pandas
import pandas as pd

# criando dicionário
data = {'peso': [50, 80, 70, 60, 65, 55, 75, 82, 61], 
     'altura': [160, 178, 175, 154, 180, 149, 177, 181, 167]}

# salvando dataframe
df = pd.DataFrame(data)

# plotar scatter (gráfico de dispersão)
sns.scatterplot(x='peso', y='altura', data = df)
plt.show()
````

### **Visualização de categorias:**
Podemos usar o ``boxplot()`` para visualizar a distribuição de uma variável categórica.


````python
# criando df
data2 = {'tipo': ['Venda', 'Venda', 'Compra', 'Compra', 'Venda', 'Compra'],
         'valor': [1000, 1500, 500, 800, 1200, 700]}

# salvando df
df2 = pd.DataFrame(data2)

# plotando 
sns.boxplot(x='tipo', y='valor', data = df2)
plt.show()
````
